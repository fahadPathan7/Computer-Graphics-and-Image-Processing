```python
import matplotlib.pyplot as plt
```

# For 0 < m < 1
eq: d1 = 2 * del(y) - del(x)

```python
def drawLineForLowM(x0, y0, x1, y1):
    points = []
    dx = x1 - x0
    dy = y1 - y0
    yi = 1
    if dy < 0:
        yi = -1
        dy = -dy
    D = 2 * dy - dx
    y = y0

    for x in range(x0, x1 + 1):
        points.append((x, y))
        if D > 0:
            y = y + yi
            D = D - 2 * dx
        D = D + 2 * dy

    return points
```

## Testing with some value (0 < m < 1)


```python
x0, y0 = 1, 1
x1, y1 = 8, 4

points_low = []

if abs(y1 - y0) < abs(x1 - x0):
    if x0 > x1:
        points_low = drawLineForLowM(x1, y1, x0, y0)
    else:
        points_low = drawLineForLowM(x0, y0, x1, y1)
```

Plotting the line

```python
x_values_low, y_values_low = zip(*points_low)

plt.plot(x_values_low, y_values_low, marker='o', label='0 < m < 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bresenham\'s Line Drawing Algorithm for 0 < m < 1')
plt.grid(True)
plt.legend()
plt.show()
```

# For m > 1
The difference from 0 < m < 1, is the operations of x will be replaced with y and the operations of y will be replaced with x.

```python
def drawLineForHighM(x0, y0, x1, y1):
    points = []
    dx = x1 - x0
    dy = y1 - y0
    xi = 1
    if dx < 0:
        xi = -1
        dx = -dx
    D = 2 * dx - dy
    x = x0

    for y in range(y0, y1 + 1):
        points.append((x, y))
        if D > 0:
            x = x + xi
            D = D - 2 * dy
        D = D + 2 * dx

    return points
```

## Testing with some value m > 1

```python
x0, y0 = 1, 1
x1, y1 = 4, 8

points_high = []

if abs(y1 - y0) > abs(x1 - x0):
    if y0 > y1:
        points_high = drawLineForHighM(x1, y1, x0, y0)
    else:
        points_high = drawLineForHighM(x0, y0, x1, y1)
```

Plotting the line

```python
x_values_high, y_values_high = zip(*points_high)

plt.plot(x_values_high, y_values_high, marker='o', label='m > 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bresenham\'s Line Drawing Algorithm for m > 1')
plt.grid(True)
plt.legend()
plt.show()
```
